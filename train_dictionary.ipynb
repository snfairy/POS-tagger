```python

# coding: utf-8

# In[2]:


import numpy as np


# In[3]:


f = open('wsj1-18.training')
contents = f.read()


# In[4]:


a = contents.split('\n')


# In[5]:



```

```python
def file_separator(a):
    content = []
    tags = []
    for i in a:
        c1 = []
        t1 = []
        sentence = list(i.split())
        start = 0
        if not len(sentence)//2:
            continue
        while start<len(sentence):
            c1.append(sentence[start])
            t1.append(sentence[start+1])
            start += 2
        content.append(c1)
        tags.append(t1)
    return [content,tags]
content,tags = file_separator(a)


# In[6]:


pi = {}
b = {}
a = {}
for i in range(len(content)):
    
    for j in range(len(content[i])):
        if content[i][j] not in b:
            b[content[i][j]] = {'total': 0}
        if tags[i][j] not in b[content[i][j]]:
            b[content[i][j]][tags[i][j]] = 1
        else:
            b[content[i][j]][tags[i][j]] += 1
        b[content[i][j]]['total'] += 1
        if j == 0:
            if tags[i][j] not in pi:
                pi[tags[i][j]] = 1
            else:
                pi[tags[i][j]]  +=1
        if j > 0:
            if tags[i][j-1] not in a:
                a[tags[i][j-1]] = {'total': 0}
            if tags[i][j] not in  a[tags[i][j-1]]:
                a[tags[i][j-1]][tags[i][j]]= 1
            else:
                a[tags[i][j-1]][tags[i][j]] += 1
            a[tags[i][j-1]]['total'] += 1


# In[8]:


for k in pi:
    pi[k]/= len(content)+1
pi['00'] = 1/(len(content)+1)


# In[18]:


for i in b:
    total = b[i]['total']
    for j in b[i]:
        if j!= 'total':
            b[i][j] /= (total + 1)
    b[i]['00'] = 1/(total+1)
    del b[i]['total']


# In[9]:


for i in a:
    total = a[i]['total']
    for j in a[i]:
        if j!= 'total':
            a[i][j] /= (total+1)
    a[i]['00'] = 1/(total+1)
    del a[i]['total']


# In[21]:


import pickle
obj = {}
obj['a'] = a
obj['pi'] = pi
obj['b'] = b


# In[22]:


def save_object(obj, filename):
    with open(filename, 'wb') as output:
        pickle.dump(obj, output, pickle.HIGHEST_PROTOCOL)
save_object(obj, 'model.pyc')
```
